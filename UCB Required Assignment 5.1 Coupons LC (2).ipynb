# Practical Application 1: Coupon Acceptance Analysis

## Business Question

The goal of this project is to analyze customer coupon acceptance behavior using Python, pandas, matplotlib, and seaborn.

This analysis compares customers who accepted coupons with customers who rejected coupons to identify patterns and trends.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Visualization 1: Overall Coupon Acceptance

```python
plt.figure(figsize=(6,4))

sns.countplot(x='Y', data=df)

plt.title('Overall Coupon Acceptance')
plt.xlabel('Accepted Coupon')
plt.ylabel('Count')

plt.xticks([0,1], ['No', 'Yes'])

plt.show()
```

## Visualization 2: Coupon Type vs Acceptance

```python
plt.figure(figsize=(10,6))

sns.countplot(x='coupon', hue='Y', data=df)

plt.title('Coupon Type vs Acceptance')
plt.xlabel('Coupon Type')
plt.ylabel('Count')

plt.xticks(rotation=45)

plt.show()
```

## Findings

After exploring the dataset and creating several visualizations, I found that coupon acceptance appeared to be influenced by multiple factors rather than one single variable. Certain coupon types, especially coffee house and lower-cost restaurant coupons, seemed to have higher acceptance rates compared to other offers. This may be because these types of purchases are more convenient, lower cost, and easier to fit into a customer’s normal routine.

The visualizations also suggested that time of day and weather conditions may affect customer behavior. In some cases, customers appeared more willing to accept a coupon when driving conditions were more favorable or when the timing aligned with common meal or coffee hours. Looking at the charts made it easier to identify patterns between customers who accepted coupons and those who did not.

As someone who is still new to working with datasets and Python, I found it interesting how visualizations can simplify large amounts of information and make trends easier to understand. Overall, the analysis suggests that convenience, timing, and the type of offer all play an important role in customer coupon acceptance behavior.

## Recommendations
